```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Unduh semua resource yang dibutuhkan
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')

# Load data hasil scraping
df = pd.read_csv("[Jobstreet]Complete_Job_Listing.csv")

# Inisialisasi stopwords dan lemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

# Fungsi preprocessing
def clean_text(text):
    if pd.isnull(text):
        return ""

    # Lowercase
    text = text.lower()

    # Hapus tautan
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)

    # Hapus hashtag dan mention
    text = re.sub(r'\@\w+|\#', '', text)

    # Hapus emoji dan karakter non-ASCII
    text = text.encode('ascii', 'ignore').decode('utf-8')

    # Hapus angka
    text = re.sub(r'\d+', '', text)

    # Hapus tanda baca
    text = text.translate(str.maketrans('', '', string.punctuation))

    # Tokenisasi
    tokens = nltk.word_tokenize(text)

    # Hapus stopwords dan lemmatization
    cleaned_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]

    # Gabungkan kembali
    return ' '.join(cleaned_tokens)

# Terapkan ke kolom 'Deskripsi'
df['Deskripsi_Cleaned'] = df['Deskripsi'].apply(clean_text)

# Simpan ke CSV baru
df.to_csv("Jobstreet_Cleaned_Descriptions.csv", index=False)

print("Preprocessing selesai dan file telah disimpan sebagai 'Jobstreet_Cleaned_Descriptions.csv'")
```

# RECOMMENDATION

**Using Content-Based Filtering** ( TF-IDF + Cosine Similarity pada kolom Deskripsi )

Content-Based Filtering menyarankan item (pekerjaan) yang memiliki kemiripan konten (dalam hal ini: deskripsi pekerjaan) dengan item yang sedang dilihat pengguna.

 Data utama yang digunakan: ***Deskripsi***

 Proses:

* Deskripsi pekerjaan diubah menjadi vektor menggunakan TF-IDF (Term Frequency–Inverse Document Frequency).

* Lalu dihitung kemiripan antar deskripsi menggunakan Cosine Similarity.

* Pekerjaan lain yang paling mirip dengan pekerjaan referensi akan ditampilkan sebagai rekomendasi

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load CSV hasil preprocessing
df = pd.read_csv("/content/Jobstreet_Cleaned_Descriptions.csv")

# Isi NaN dengan string kosong
df['Deskripsi_Cleaned'] = df['Deskripsi_Cleaned'].fillna('')

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Deskripsi_Cleaned'])

# Hitung cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Fungsi untuk mendapatkan rekomendasi
def get_recommendations(title, df=df, cosine_sim=cosine_sim):
    # Buat index pencarian lowercase dari kolom Posisi
    indices = pd.Series(df.index, index=df['Posisi'].str.lower())

    title = title.lower()

    # Check if the title exists in the index
    if title not in indices.index:
        return f"Posisi '{title}' tidak ditemukan di data."

    # Get the index for the title (handle potential duplicates by taking the first)
    idx = indices[title].iloc[0] if isinstance(indices[title], pd.Series) else indices[title]

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:235]

    job_indices = [i[0] for i in sim_scores]
    return df[['Posisi', 'Perusahaan', 'Lokasi', 'Gaji', 'Waktu', 'Deskripsi_Cleaned']].iloc[job_indices]


# Contoh pemakaian
rekomendasi = get_recommendations("Data Scientist")
print(rekomendasi)
```

**Recommendation with Similarity Score**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Load CSV hasil preprocessing
df = pd.read_csv("/content/Jobstreet_Cleaned_Descriptions.csv")

# Isi NaN dengan string kosong
df['Deskripsi_Cleaned'] = df['Deskripsi_Cleaned'].fillna('')

# TF-IDF Vectorizer
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['Deskripsi_Cleaned'])

# Hitung cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Fungsi untuk mendapatkan rekomendasi
def get_recommendations(title, df=df, cosine_sim=cosine_sim):
    # Buat index pencarian lowercase dari kolom Posisi
    indices = pd.Series(df.index, index=df['Posisi'].str.lower())

    title = title.lower()

    # Cek apakah title ada di dalam index
    if title not in indices.index:
        return f"Posisi '{title}' tidak ditemukan di data."

    # Ambil index-nya
    idx = indices[title].iloc[0] if isinstance(indices[title], pd.Series) else indices[title]

    # Hitung skor similarity
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:235]  # Ambil top rekomendasi (tanpa dirinya sendiri)

    job_indices = [i[0] for i in sim_scores]
    similarity_scores = [i[1] for i in sim_scores]

    # Ambil data dan tambahkan skor similarity
    result_df = df[['Posisi', 'Perusahaan', 'Lokasi', 'Gaji', 'Waktu', 'Deskripsi_Cleaned']].iloc[job_indices].copy()
    result_df['Similarity_Score'] = similarity_scores

    return result_df

# Contoh pemakaian
rekomendasi = get_recommendations("Data Scientist")
print(rekomendasi)
```
